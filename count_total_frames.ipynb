```python
def time_sec(x):
    try:        
        x = str(x)
        minute = int(x.split('.')[0])*60
        sec = int(x.split('.')[1])
        time = minute+sec
        return time
        
    except:
        pass
```

```python
data_total['from_sec'] = data_total['From (min.sec)'].apply(time_sec)
```

```python
data_total['to_sec'] = data_total['To (min.sec)'].apply(time_sec)
```
