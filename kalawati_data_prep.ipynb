```python
def localize(x):
    x = str(x)
    return pd.to_datetime((x).split("+")[0])
```

```python
df.starttime = df.starttime.apply(localize)
```

```python
ds = pd.DataFrame()
```

```python
for i in data['Video (Title)'].unique():
    print(i)
    patients = data[data['Video (Title)']==i]
    title = i
    d = patients['uhid'].iloc[0]
    #print(d)
    for label in patients.Label.unique():              
        l = label
        p = patients[patients['Label']==label]    
        
        start_video = pd.to_datetime(p.START.iloc[0]) - timedelta(seconds=1200) + timedelta(seconds=19800)
        end_video = pd.to_datetime(p.END.iloc[0]) + timedelta(seconds=1200) + timedelta(seconds=19800)
        q = df[(df['uhid']==str(d)) & (df['starttime']>=start_video) & (df['starttime']<=end_video)]
        #q = df[df['uhid']==str(d)]
        print(q)
        q['label'] = l
        q['procedure_start'] = pd.to_datetime(p.START.iloc[0])
        q['procedure_end'] = pd.to_datetime(p.END.iloc[0])
        q['title'] = title
        ds = ds.append(q,ignore_index=True)
    
    
    
        
    
```

```python
ds.to_csv('kalawati_prepped_data_6.csv')
```
