```python
#Null Hypothesis : The variances in tbe two groups is not different (pre and during)
#Alternate Hypothesis : The variances in tbe two groups is different (pre and during)
```

```python
HR_pre = df.dropna(subset=['baseline_HR','during_HR'],axis=0)
```

```python
stat_levene_HR, p_value_levene_HR = levene(HR_pre['baseline_HR'],HR_pre['during_HR'])
```

```python
#The Levene test reveals that the variances of the two groups are statistically not different as the P value is above 0.05 at 95% Confidence
print(stat_levene_HR,p_value_levene_HR)
```

```python
#Null Hypothesis : The variances in tbe two groups is not different (post and during)
#Alternate Hypothesis : The variances in tbe two groups is different (post and during)
```

```python
HR_post = df.dropna(subset=['post_HR','during_HR'],axis=0)
```

```python
stat_levene_HR, p_value_levene_HR = levene(HR_post['post_HR'],HR_post['during_HR'])
```

```python
#The Levene test reveals that the variances of the two groups are statistically not different as the P value is above 0.05 at 95% Confidence
print(stat_levene_HR,p_value_levene_HR)
```

```python
#Null hypothesis: HR for pre and during is not different
#Alternate Hypothesis: HR is different
HR_t_stat, HR_p_val = ttest_ind(HR_pre['baseline_HR'],HR_pre['during_HR'], equal_var = True)
```

```python
#Looking at the T Statistics and P value, we cannot reject the null hypothesis
#The HR for pre and during procedure is not different as p value is above 0.05
print(HR_t_stat,HR_p_val)
```

```python
#Null hypothesis: HR for post and during is not different
#Alternate Hypothesis: HR is different
HR_t_stat, HR_p_val = ttest_ind(HR_post['post_HR'],HR_post['during_HR'], equal_var = True)
```

```python
#Looking at the T Statistics and P value, we cannot reject the null hypothesis
#The HR for post and during procedure is not different as p value is above 0.05
print(HR_t_stat,HR_p_val)
```

```python
#Conclusion: The Heartrate pre, post and during procedure does not show significant difference.
```
