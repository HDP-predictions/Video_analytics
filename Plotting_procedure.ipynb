```python
ti = 'NEO101.stream_720p_2020-01-06-13.01.28.782-UTC_1.mp4'
```

```python
q = x[(x['title']==ti)&(x['label']=='Patting')]
```

```python
q.sort_values('starttime',inplace=True)
```

```python
#q
```

```python
q['rolling_HR'] = q.heartrate.rolling(10).mean()
q['rolling_Spo2'] = q.spo2.rolling(10).mean()
```

```python
q = q.dropna(subset=['rolling_HR','rolling_Spo2'])
```

```python
q['HR_residual'] = q['heartrate'] - q['rolling_HR']
q['spo2_residual'] = q['spo2'] - q['rolling_Spo2']
```

```python
end_video = pd.to_datetime(q.procedure_end.iloc[0]) + timedelta(seconds=600) + timedelta(seconds=19800)+timedelta(seconds=120)
```

```python
start_video = pd.to_datetime(q.procedure_start.iloc[0]) + timedelta(seconds=19800)-timedelta(seconds=120)
end = pd.to_datetime(q.procedure_end.iloc[0])+ timedelta(seconds=19800)+timedelta(seconds=120)
```

```python
q = q[pd.to_datetime(q['starttime'])<end_video]
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20, 6
axes = plt.gca()

ax = plt.subplot()
sns.lineplot(y = (q['heartrate']), x = (q['starttime']),label='HR')
sns.lineplot(y = (q['rolling_HR']), x = (q['starttime']),label='Rolling HR')
sns.lineplot(y = (q['HR_residual']), x = (q['starttime']),label='Residual HR')
plt.xticks(rotation=70)
#plt.axvline(str(end))
#sns.lineplot(y = (ecg_resprate_1['ecg_resprate']), x = np.arange(len((ecg_resprate_1['ecg_resprate']))),label='RR')
plt.legend(loc=0)
ax.legend()
plt.savefig('heartrate_patting_3.png',dpi=300)
```
